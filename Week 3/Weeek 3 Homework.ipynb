```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.2, random_state=42)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.y.values
y_val = df_val.y.values
y_test = df_test.y.values

del df_train['y']
del df_val['y']
del df_test['y']
```

```python
df_categories = df[["job","marital","education","housing","loan","contact","month","poutcome"]]
```

```python
df_categories.columns
```

```python
from sklearn.metrics import mutual_info_score
```

```python
for category in df_categories.columns:
    print("Categoty",category,": ", round(mutual_info_score(df_full_train.y, df_full_train[category]),2))
```

```python
categorical = ["job","marital","education","housing","loan","contact","month","poutcome"]
numerical = ["age", "balance", "day", "duration", "campaign", "pdays", "previous"]
```

```python
from sklearn.feature_extraction import DictVectorizer
train_dict = df_train[categorical + numerical].to_dict(orient='records')
train_dict[0]
```

```python
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
X_train.shape
```

```python
# dv.get_feature_names()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
model.predict_proba(X_val)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
```

```python
y_pred
```

```python
churn = y_pred > 0.5
```

```python
round((y_val == churn).mean(),2)
```
